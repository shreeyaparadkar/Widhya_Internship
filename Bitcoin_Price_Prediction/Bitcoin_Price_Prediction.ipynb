#### Microtask - 4: Data Cleaning

```python
dfnew=df[['btc_market_cap', 'btc_n_transactions','btc_miners_revenue','btc_cost_per_transaction','btc_difficulty','btc_hash_rate','btc_cost_per_transaction_percent','btc_market_price']]
dfnew.head()
```

```python
dfnew.info()
```

```python
dfnew.isna().any() #identifying null values
```

```python
dfnew.isna().sum() #count of null values
```

```python
dfnew.btc_difficulty.describe()
```

```python
dfnew_median=dfnew.btc_difficulty.median() 
dfnew_median
```

```python
dfnew.btc_difficulty.fillna(dfnew_median, inplace=True) #replacing null values with mean
```

```python
dfnew.isna().any() 
```

#### Microtask - 5: Model Formulation & Prediction

```python
X=dfnew.iloc[:,:-1].values
Y=dfnew.iloc[:,1].values
```

#### Train-Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.25, random_state=0) 
```

#### Model Building

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,Y_train)
```

#### Training-Testing Score

```python
print('Training Score: ', round(lr.score(X_train,Y_train)*100,2),'%')
print('Testing Score: ', round(lr.score(X_test,Y_test)*100,2))
```

#### Prediction

```python
Y_predict=lr.predict(X_test)
Y_predict
```

```python
Y_test
```

```python
from sklearn import metrics
acc=metrics.r2_score(Y_test,Y_predict)
print("Accuracy Score of Model: ",round(acc*100,2),'%')
```

#### Actual Price v/s Predicted Price

```python
prediction_df = pd.DataFrame({ 'Actual Price': Y_test, 'Predicted Price': Y_predict})  
prediction_df
```

#### Model Evaluation

```python
from sklearn import metrics
print('Mean Absolute Error:',round(metrics.mean_absolute_error(Y_test,Y_predict),2))
print('Mean Squared Error:',round(metrics.mean_squared_error(Y_test,Y_predict),2))
print('Root Mean Squared Error:',round(np.sqrt(metrics.mean_squared_error(Y_test,Y_predict)),2))
print('Explained Variance Score:',metrics.explained_variance_score(Y_test,Y_predict))
```
