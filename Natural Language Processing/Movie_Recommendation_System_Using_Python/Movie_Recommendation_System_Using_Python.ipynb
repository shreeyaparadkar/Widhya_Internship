#### Importing Libraries

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

#### Reading Data

```python
df = pd.read_csv('movie.csv')
```

#### Selecting required features

```python
features = ['keywords', 'cast', 'genres', 'director']
```

```python
features
```

#### Replacing NaN values with an empty string

```python
for feature in features:
    df[feature] = df[feature].fillna('')
```

#### Creating Function: 'combined_features'

```python
def combined_features(row):
    return row['keywords']+" "+row['cast']+" "+row['genres']+" "+row['director']
```

```python
df["combined_features"] = df.apply(combined_features, axis =1)
```

```python
df.head()
```

#### Vectorizing data

```python
cv=CountVectorizer()
count_vector=cv.fit_transform(df['combined_features'])
```

```python
print("Count Vector:", count_vector.toarray())
```

#### Microtask - 2: What is the Similarity Score of a movie with itself?

```python
x=[]
y=0
for i in count_vector:
    x.append([cosine_similarity(i,count_vector[0]),y])
    y+=1
```

```python
x[0]
```

```python
cosine_sim = cosine_similarity(count_vector)
print(cosine_sim)
```

> The similarity score of a movie with itself is 1

#### Creating a function to identify similar movies
